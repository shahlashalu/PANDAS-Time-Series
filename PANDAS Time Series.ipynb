```python
dates = ['2022-11-12 2:30:00 PM', 'jan 5,2024 18:44:00', '01/11/2021', '2020.01.11', '2022/01/11', '20200105']

import pandas as pd

date_time = pd.to_datetime(dates)

date_time
```

```python
import pandas as pd

dates = '11-january-24 12:30:00'

date_time = pd.to_datetime(dates)

print(date_time)

print('\n')

print(date_time.day)
print('\n')
print(date_time.week)
print('\n')
print(date_time.day_of_week)
print('\n')
print(date_time.day_of_year)
print('\n')
print(date_time.minute)
print('\n')
print(date_time.is_leap_year)
print('\n')
print(date_time.is_month_end)
print('\n')
print(date_time.is_month_start)
print('\n')
print(date_time.month)
print('\n')
print(date_time.quarter)
```
